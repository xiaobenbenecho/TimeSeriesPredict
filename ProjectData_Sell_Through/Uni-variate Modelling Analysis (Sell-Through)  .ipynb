## Model Performance on Sell-through PC Analysis

```python
#Jupyter notebook
import warnings
warnings.filterwarnings('ignore')
import pandas as pd 
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 10]
```

## Model comparison fo PC 

```python
#Read in the three tables
table1 = pd.read_pickle('C:\\Users\\yzhang\\Desktop\\KTPAssociate\\Data\\ResultsProjectSellThroughPC\\df_inclusive_normal_W6.pkl')
table2 = pd.read_pickle('C:\\Users\\yzhang\\Desktop\\KTPAssociate\\Data\\ResultsProjectSellThroughPC\\df_inclusive_outliers_interpolated_mean_W6.pkl')
table3 = pd.read_pickle('C:\\Users\\yzhang\\Desktop\\KTPAssociate\\Data\\ResultsProjectSellThroughPC\\df_inclusive_outliers_interpolated_iqr_W6.pkl')
```

Results: Comparison of different pre-processing methods

The boxplots showing the prediction errors of the four models without any pre-processing, with outliers interpolated by mean, with outliers interpolated by the nearest bound.

```python
table1['window_size'].value_counts()
```

```python
table2['window_size'].value_counts()
```

```python
table3['window_size'].value_counts()
```

```python
cols = [col for col in table1.columns if 'pe' in col]
cols
```

```python
value1 = np.log(table1[cols].apply(pd.to_numeric))
value2 = np.log(table2[cols].apply(pd.to_numeric))
value3 = np.log(table3[cols].apply(pd.to_numeric))

#value1 = table1[cols].apply(pd.to_numeric)
#value2 = table2[cols].apply(pd.to_numeric)
#value3 = table3[cols].apply(pd.to_numeric)
cols = ['ARMA', 'Prophet','MTM','MTMRW','Naive','Average','SARIMAX','UC']
value1.columns = cols
value2.columns = cols
value3.columns = cols
```

```python
plt.rcParams['figure.figsize'] = [14, 10]
f, axes = plt.subplots(1,3)
plt.tight_layout()
sns.boxplot(x="variable", y="value", data=pd.melt(value1), ax = axes[0]).set_title("Without Pre-Processing")
sns.boxplot(x="variable", y="value", data=pd.melt(value2), ax = axes[1]).set_title("Outliers Interpolated by Mean")
sns.boxplot(x="variable", y="value", data=pd.melt(value3), ax = axes[2]).set_title("Outliers Interpolated by Nearest Boundary")
```
